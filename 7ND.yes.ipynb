### 20. Sukurkite funkcija is_tropical() winemag-data.csv duomenu rinkinyje,
### kuri priima string kintamaji ir grazina ar jame yra zodis 'tropical' (True/False)

```python
def is_tropical(row):
    if 'tropical' in row:
        return True
    else:
        return False
```

```python
is_tropical('tropical')
```

```python
# 21. 

reviews['description_words'] = reviews['description'].map(is_tropical)
reviews['description_words'].sum()
```
