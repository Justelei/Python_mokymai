```python
# 1. Is pirmu 100 skaiciu, jeigu skaicius dalinasi is 2 be liekanos, atspausdinkite 'Fizz',
# jeigu skaicius dalinasi is 3 be lieknos, atspausdinkite 'Buzz',
# jei skaicius dalinasi is 2 ir 3 be liekanos, atspausdinkite 'FizzBuzz'



for number in range (1,101):
    if (number % 6) == 0:
        print (number, 'FizzBuzz')
    elif (number % 2) == 0:
        print(number, 'Fizz')
    elif (number % 3) == 0:
        print (number, 'Buzz')
    else:
        print (number, 'number is not dividible by 2, 3 or 6')
```

```python
for number in range (1,101):
    result_text = ''
    if number % 2 == 0:
        result_text += 'Fizz'
    if number % 3 == 0:
        result_text += 'Buzz'
    print(number, result_text)
```

```python
# Leiskite vartotojui pasirinkti skaiciu, jeigu ivestas sakicius yra daugiau arba lygus 10,
# atspausdinkite visus skaicius, kurie dalinasi is 5 su liekana,
# kitu atveju nieko nedaryti ir pranesti vartotojui apie tai.

input_number = int(input('Iveskite skaiciu'))
print(input_number)


if input_number >= 10:
    liekana = []
    for number in range(1, input_number + 1):
        if number % 5 != 0:
            liekana.append(number)
    print(liekana)
else:
    print ('pasirinktas skaicius maziau nei 10')
```

```python
# 3. Sudarykite visas imanomas kombinacijas is pateiktu 3 listu.
# Kiek yra is viso kombinacju?
# Isrikiuokite kombinaciju sarasa nuo didziausios iki maziausios reiksmes.

a = ['1', '2', '3', '4', '5']
b = ['labas', 'vakaras', 'kokia grazi diena', 'siandien uz lango', '10 laipsniu silumos']
c = ['100', '200', 40, 10]
```
