```python
for number in range(1,101):
    if number % 6 == 0:
        print(number,'FizzBuzz')
    elif number % 3 == 0:
        print(number,'Buzz')
    elif number % 2 == 0:
        print(number, 'Fizz')
```

```python
for numbers in range(1,101):
    result_text = ''
    if numbers % 2 == 0:
        result_text += 'Fizz'
    if numbers % 3 == 0:
        result_text +='Buzz'
    print(numbers, result_text)
    
```

```python
array_2d = np.array( [ [1,2], [4,3], [7,4] ] )
print(array_2d)
print(array_2d.shape)

array_3d = np.array( [
    [ [1,2], [4,3], [7,4] ],
    [ [1,2], [4,3], [7,4] ],
    [ [1,2], [4,3], [7,4] ],
    [ [1,2], [4,3], [7,4] ],
]
)
     
print('3D\n', array_3d)
print(array_3d.shape)
```

# numpy methods

```python
np.mean(array_3d)
```

```python
np.mean(array_3d, axis=0)
```

```python
print(array_2d)
print('Column mean', np.mean(array_2d, axis=0))
print('Row mean', np.mean(array_2d, axis=1))
```

```python
np.std(array_2d)
print('Column std', np.std(array_2d, axis=0))
```

```python
np.sum(array_2d)
print('Column sum', np.sum(array_2d, axis=0))
print('Row sum', np.sum(array_2d, axis=1))
```
